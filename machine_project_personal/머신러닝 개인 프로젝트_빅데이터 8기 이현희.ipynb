## 머신러닝 개인 프로젝트

```python
import pandas as pd  
import numpy as np

train_data = pd.read_csv('../hyunhee/learning/machinelearning/train.csv')
test_data = pd.read_csv('../hyunhee/learning/machinelearning/test.csv')
```

```python
test_data
```

```python
binary_mapping = {
    'school': {'GP': 0, 'MS': 1},
    'sex': {'F': 0, 'M': 1},
    'address': {'U': 0, 'R': 1},
    'famsize': {'LE3': 0, 'GT3': 1},
    'Pstatus': {'T': 0, 'A': 1},
    'schoolsup': {'yes': 1, 'no': 0},
    'famsup': {'yes': 1, 'no': 0},
    'paid': {'yes': 1, 'no': 0},
    'activities': {'yes': 1, 'no': 0},
    'nursery': {'yes': 1, 'no': 0},
    'higher': {'yes': 1, 'no': 0},
    'internet': {'yes': 1, 'no': 0},
    'romantic': {'yes': 1, 'no': 0}
}

for column, mapping in binary_mapping.items():
    train_data[column] = train_data[column].map(mapping)
    test_data[column] = test_data[column].map(mapping)
```

```python
# 원-핫 인코딩 (True = 1 , False = 0)
categorical_columns = ['Fjob', 'Mjob', 'reason', 'guardian']
train_data_encoded = pd.get_dummies(train_data, columns=categorical_columns, dtype=int)
test_data_encoded = pd.get_dummies(test_data, columns=categorical_columns, dtype=int)
```

```python
data_input = train_data_encoded[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid',
       'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel',
       'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2',
       'Fjob_at_home', 'Fjob_health', 'Fjob_other', 'Fjob_services',
       'Fjob_teacher', 'Mjob_at_home', 'Mjob_health', 'Mjob_other',
       'Mjob_services', 'Mjob_teacher', 'reason_course', 'reason_home',
       'reason_other', 'reason_reputation', 'guardian_father',
       'guardian_mother', 'guardian_other']].to_numpy()

target = train_data_encoded['G3'].to_numpy()
```

```python
correlations = np.corrcoef(data_input.T, target)
```

```python
data_input = train_data_encoded[['school', 'sex', 'age', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'traveltime', 'studytime', 'failures', 'schoolsup', 'famsup', 'paid',
       'activities', 'nursery', 'higher', 'internet', 'romantic', 'famrel',
       'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences', 'G1', 'G2',
       'Fjob_at_home', 'Fjob_health', 'Fjob_other', 'Fjob_services',
       'Fjob_teacher', 'Mjob_at_home', 'Mjob_health', 'Mjob_other',
       'Mjob_services', 'Mjob_teacher', 'reason_course', 'reason_home',
       'reason_other', 'reason_reputation', 'guardian_father',
       'guardian_mother', 'guardian_other']].to_numpy()

target = train_data_encoded['G3'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(data_input, target, test_size= 0.2, random_state = 42)
```

```python
sub_input, val_input, sub_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state = 42)
```

```python
print(sub_input.shape, val_input.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(sub_input, sub_target)

print(rf.score(sub_input, sub_target))
print(rf.score(val_input, val_target))
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(rf, sub_input, sub_target, return_train_score=True)
print(scores)
```

```python
import numpy as np

print(np.mean(scores['train_score']))
print(np.mean(scores['test_score']))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth':range(20,50),
    'min_samples_split': range(2,25),
    'min_samples_leaf': range(1, 25),
    'min_impurity_decrease' : np.arange(0.0001, 0.01, 0.0005)
}
```

```python
gs = GridSearchCV(RandomForestRegressor(random_state = 42) , params, n_jobs=-1)
gs.fit(sub_input, sub_target) 
```

```python
# 그리드 서치를 통해 얻은 최적의 모델(여기서는 결정 트리)
rf = gs.best_estimator_
# 최적의 모델에 대해 학습 데이터에서의 정확도를 출력
print(rf.score(sub_input, sub_target))
```

```python
print(rf.best_params_)
```

```python
rf =RandomForestRegressor(max_depth=20, min_impurity_decrease= 0.0001, min_samples_leaf= 8, min_samples_split= 2, random_state=42)
rf.fit(sub_input, sub_target)

print(rf.score(sub_input, sub_target))
print(rf.score(val_input, val_target))
```

```python
cv_results = cross_validate(rf, sub_input, sub_target, cv=5, scoring='r2', return_train_score=True)

print("Train scores:", cv_results['train_score'])
print("Validation scores:", cv_results['test_score'])
print("Average train score:", np.mean(cv_results['train_score']))
print("Average validation score:", np.mean(cv_results['test_score']))
```
